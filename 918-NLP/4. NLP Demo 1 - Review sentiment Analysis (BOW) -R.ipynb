```r
fun <- function(str){
    if (str == 'happy'){
        return(0)
    }
    else{
        return(1)
    }
}
```

```r
data['target'] <- sapply(data$Is_Response,fun)
```

### Create test data

```r
corpus <- Corpus(VectorSource(test$Description))
str(corpus)
length(corpus$content)
corpus <- tm_map(corpus, tolower)
corpus <- tm_map(corpus, removePunctuation)
corpus <- tm_map(corpus, removeNumbers)
corpus <- tm_map(corpus, removeWords, stopwords("english"))
tdm<-TermDocumentMatrix(corpus)

TDM.common = removeSparseTerms(tdm, 0.99)
dim(TDM.common)
dim(tdm)
inspect(TDM.common)
M=as.matrix(TDM.common)#convert the bigmemory object again to a regular matrix
dim(test)
Y_test = (test$target)
X_test <-t(M)

df.test <- data.frame(X_test)
class(df.test)

```

```r
fitted.probabilities <- predict(log.model,newdata=df.test,type='response')
```

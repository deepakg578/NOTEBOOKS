```python
train.drop(['Ad Topic Line','City','Country','Timestamp'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(train.drop('Clicked on Ad',axis=1),train['Clicked on Ad'],
                                                 test_size=0.30,random_state=101 )
```

```python
X_train.info()
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, auc ,roc_curve
```

```python
print(len(y_test),sum(y_test))
```

```python
print(classification_report(y_test,predictions))
```
